1. **Pseudocode**

1. Siapkan Catatan Jarak: Buat sebuah daftar untuk mencatat jarak dari titik awal ke semua titik lainnya. Awalnya, isi semua dengan nilai "Tak Terhingga", kecuali titik awal yang diisi dengan 0.

2. Siapkan Antrean: Masukkan titik awal ke dalam antrean pemeriksaan.

3. Proses Titik Terdekat: Selama antrean masih ada isinya, ambil satu titik yang memiliki jarak paling kecil untuk diperiksa tetangganya.

4. Cek Tetangga: Untuk setiap titik tetangga yang terhubung:
Hitung total jarak baru (Jarak titik sekarang + bobot jalan ke tetangga).
Jika jarak baru ini lebih kecil dari jarak yang tercatat sebelumnya, maka update catatan jaraknya dengan angka yang baru.
Masukkan tetangga tersebut ke dalam antrean untuk diperiksa nanti.

5. Selesai: Jika antrean kosong, daftar catatan jarak tadi adalah hasil rute terpendek yang paling efisien.

2. **Coding**

```python
import heapq

def dijkstra(graph, start):
    # Inisialisasi jarak semua simpul sebagai tak terhingga
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0

    # Priority queue untuk menyimpan (jarak, simpul)
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Jika jarak di antrean lebih besar, abaikan
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # Jika ditemukan jalur yang lebih pendek
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Contoh Graf (Representasi Adjacency List)
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3}
}

# Jalankan fungsi
hasil = dijkstra(graph, 'A')
print("Jarak terpendek dari titik A ke titik lainnya:")
for node, dist in hasil.items():
    print(f"Ke {node} : {dist}")
```

3. **Analisa Kompleksitas Waktu**

Kompleksitas: $O((V + E) \log V)$.

V (Vertex): Jumlah titik dalam graf.
  
E (Edge): Jumlah garis atau hubungan antar titik.

Log V: Berasal dari efisiensi penggunaan Priority Queue (heap) saat mengambil nilai terkecil dan memasukkan data baru.
