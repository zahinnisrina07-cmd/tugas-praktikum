**Analisis Kompleksitas Algoritma: Optimasi dan Trade-off**

**1. Trade-off Waktu dan Ruang (Time vs Space Trade-off)**

Trade-off adalah pertukaran antara penggunaan waktu eksekusi dan memori.

Menghemat waktu biasanya membutuhkan memori lebih besar.
Menghemat memori biasanya membutuhkan waktu lebih lama.

Contoh:

Memoization menggunakan memori tambahan untuk mempercepat proses.
Iterasi biasa menggunakan memori kecil tetapi bisa lebih lambat.

**2. Teknik Optimasi Umum**

**A. Memoization**

Memoization menyimpan hasil perhitungan sebelumnya agar tidak dihitung ulang.

```python
# Contoh Tanpa Memoization (Rekursif Fibonacci)

def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

# Driver Program
n = 10
print("Fibonacci:", fib_recursive(n))
```

Analisis Kompleksitas:

Persamaan rekursi:

T(n)=T(n-1)+T(n-2)+1

Kompleksitas waktu:

O(2^n)

Karena banyak pemanggilan fungsi berulang.

**Contoh Dengan Memoization**

```python
# Contoh Dengan Memoization
# Fibonacci Memoization
memo = {}

def fib_memo(n):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib_memo(n-1) + fib_memo(n-2)
    return memo[n]

# Driver Program
n = 50
print("Fibonacci:", fib_memo(n))
```

Analisis Kompleksitas:

Setiap nilai dihitung satu kali.

Kompleksitas waktu:

O(n)

Kompleksitas ruang:

O(n)

**B. Dynamic Programming (DP)**

Dynamic Programming memecah masalah menjadi submasalah dan menyimpan hasilnya.

```python
# DP Bottom-Up Fibonacci
# Dynamic Programming Fibonacci
def fib_dp(n):

    if n <= 1:
        return n

    dp = [0] * (n + 1)

    dp[0] = 0
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

# Driver Program
n = 50
print("Fibonacci DP:", fib_dp(n))
```

Analisis Kompleksitas:

Loop berjalan sebanyak n kali.

Kompleksitas waktu:

O(n)

Kompleksitas ruang:

O(n)

**Optimasi Space DP**

DP dapat dioptimalkan karena hanya membutuhkan dua nilai sebelumnya.

```python
# Space Optimized DP Fibonacci
def fib_optimized(n):

    if n <= 1:
        return n

    a, b = 0, 1

    for i in range(2, n + 1):
        a, b = b, a + b

    return b

# Driver Program
n = 50
print("Optimized Fibonacci:", fib_optimized(n))
```

Analisis Kompleksitas:

Kompleksitas waktu:

O(n)

Kompleksitas ruang:

O(1)

**C. Greedy Algorithm**

Greedy memilih solusi terbaik lokal pada setiap langkah.

Contoh Greedy Coin Change:

```python
# Greedy Coin Change
def coin_change_greedy(coins, amount):

    # Urutkan dari terbesar
    coins.sort(reverse=True)

    result = []

    for coin in coins:

        while amount >= coin:
            amount -= coin
            result.append(coin)

    return result

# Driver Program
coins = [100, 50, 25, 10, 5, 1]
amount = 289

hasil = coin_change_greedy(coins, amount)

print("Koin digunakan:", hasil)
print("Jumlah koin:", len(hasil))
```

Analisis Kompleksitas:

n = jumlah jenis koin
k = jumlah koin yang dipilih

Kompleksitas waktu:

O(n + k)

Kompleksitas ruang:

O(k)

**Coin Change Algorithm**

Coin Change Algorithm digunakan untuk menentukan kombinasi koin agar mencapai jumlah tertentu.

Digunakan pada:

- Sistem pembayaran ATM
- E-wallet
- Optimasi kombinasi
- Dynamic Programming

```python
# Coding DP Combination
def count_ways(coins, amount):

    dp = [0] * (amount + 1)

    dp[0] = 1

    for coin in coins:

        for i in range(coin, amount + 1):
            dp[i] += dp[i-coin]

    return dp[amount]

# Driver Program
coins = [1, 2, 5]
amount = 5

print("Jumlah kombinasi:", count_ways(coins, amount))
```

Analisis Kompleksitas:

Kompleksitas waktu:

O(n × A)

Kompleksitas ruang:

O(A)
