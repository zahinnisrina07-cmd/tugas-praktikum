Analisis Kompleksitas Waktu

**Pseudocode**

1. Mulai
2. Masukkan nilai n
3. Inisialisasi hasil = 1
4. Untuk i dari 1 sampai n lakukan:
5.     hasil = hasil × i
6. Tampilkan hasil
7. Selesai

```python
def hitung_faktorial(n):
    total = 1
    for i in range(1, n + 1):
        total *= i
    return total

angka = 6
print("Hasil faktorial dari", angka, "adalah", hitung_faktorial(angka))
```

**Analisis Kompleksitas**

Program ini menghitung nilai faktorial dengan menggunakan perulangan dari 1 hingga n.

Setiap iterasi melakukan satu operasi perkalian, sehingga jumlah langkah yang dijalankan sebanding dengan nilai n.

Karena hanya terdapat satu perulangan utama, maka kompleksitas waktu dari algoritma ini adalah O(n).
