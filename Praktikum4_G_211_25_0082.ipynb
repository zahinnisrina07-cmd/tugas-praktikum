**1. Linear Search**

Pencarian linear adalah metode pencarian yang membandingkan setiap elemen array satu per satu dari awal hingga akhir.

```python
# [Linear Search]
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Contoh
data = [10, 20, 30, 40, 50]
print(f"Hasil Linear Search: {linear_search(data, 30)}")
```

Analisa Kompleksitas & Case:

Kompleksitas Waktu Big O: $O(n)$

Best Case: $O(1)$ — Target berada di indeks pertama.

Average Case: $O(n)$ — Target berada di posisi acak atau di tengah.

Worst Case: $O(n)$ — Target berada di akhir atau tidak ditemukan sama sekali.

**2. Binary Search (Iteratif)**

Mencari dengan cara membagi dua area pencarian terus-menerus. Syarat utamanya adalah data harus sudah terurut.

```python
# [Binary Search Iteratif]
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Contoh
data = [10, 20, 30, 40, 50]
print(f"Hasil Binary Search: {binary_search(data, 40)}")
```

Analisa Kompleksitas & Case:

Kompleksitas Waktu Big O: $O(\log n)$

Best Case: $O(1)$ — Target adalah nilai tengah pertama yang diperiksa.

Average Case: $O(\log n)$ — Area data berkurang setengah di setiap langkah.

Worst Case: $O(\log n)$ — Target ditemukan pada pembagian terakhir atau tidak ditemukan.

**3. Binary Search (Rekursif)**

Logikanya sama dengan iteratif, namun menggunakan pemanggilan fungsi ke dirinya sendiri.  

```python
# [Binary Search Rekursif]
def binary_search_recursive(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, mid + 1, right, target)
    else:
        return binary_search_recursive(arr, left, mid - 1, target)

# Contoh
data = [10, 20, 30, 40, 50]
print(f"Hasil Binary Search Rekursif: {binary_search_recursive(data, 0, len(data)-1, 20)}")
```

Analisa Kompleksitas & Case:

Kompleksitas Waktu Big O: $O(\log n)$

Best Case: $O(1)$ — Ditemukan langsung pada pemanggilan fungsi pertama.

Average Case: $O(\log n)$ — Mengikuti pola pembagian logaritmik.

Worst Case: $O(\log n)$ — Rekursi mencapai batas terdalam tanpa menemukan target.

**4. Jump Search**

Melompati elemen dengan jarak blok tetap (misal $\sqrt{n}$) untuk mencari blok di mana target berada

```python
import math

# [Jump Search]
def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1
    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i
    return -1

# Contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(f"Hasil Jump Search: {jump_search(data, 60)}")
```

Analisa Kompleksitas & Case:

Kompleksitas Waktu Big O: $O(\sqrt{n})$

Best Case: $O(1)$ — Target adalah elemen pertama array.

Average Case: $O(\sqrt{n})$ — Rata-rata lompatan dan langkah linear yang dibutuhkan.

Worst Case: $O(\sqrt{n})$ — Target di akhir blok terakhir atau tidak ditemukan.

**5. Interpolation Search**

Mencari posisi target menggunakan rumus estimasi posisi berdasarkan nilai data.

```python
# [Interpolation Search]
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high and target >= arr[low] and target <= arr[high]:
        if low == high:
            return low if arr[low] == target else -1
        pos = low + ((target - arr[low]) * (high - low) // (arr[high] - arr[low]))
        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1
    return -1

# Contoh
data = [10, 20, 30, 40, 50]
print(f"Hasil Interpolation Search: {interpolation_search(data, 50)}")
```

Analisa Kompleksitas & Case:

Kompleksitas Waktu Big O: $O(n)$ (Worst Case) / $O(\log(\log n))$ (Average Case).

Best Case: $O(1)$ — Rumus interpolasi langsung menebak posisi dengan tepat.

Average Case: $O(\log(\log n))$ — Jika distribusi data sangat merata/seragam.

Worst Case: $O(n)$ — Jika distribusi data sangat tidak merata (mirip linear search).
