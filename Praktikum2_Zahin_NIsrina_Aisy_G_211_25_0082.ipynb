Red-Black Tree dan Analisis Kompleksitas

**1. Pseudocode Algoritma Red-Black Tree**

1. Insert: Masukkan data baru seperti BST biasa dan beri warna MERAH.

2. Check Root: Jika node baru adalah Root, ubah warnanya jadi HITAM.

3. Fix-up Rule: Jika Parent berwarna MERAH, lihat warna Paman (saudara Parent):
Kasus 1: Jika Paman MERAH, ubah Parent & Paman jadi HITAM, lalu Kakek jadi MERAH.
Kasus 2: Jika Paman HITAM dan posisi node membentuk sudut, lakukan Rotasi pada Parent.
Kasus 3: Jika Paman HITAM dan posisi node lurus, ubah warna Parent jadi HITAM, Kakek jadi MERAH, lalu Rotasi pada Kakek.

4. Final: Pastikan Root tetap HITAM.

```python
class Node:
    def __init__(self, data, color=1):
        self.data = data
        self.color = color  # 1: Merah, 0: Hitam
        self.left = self.right = self.parent = None

class SimpleRBT:
    def __init__(self):
        self.nil = Node(0, 0)
        self.root = self.nil

    def rotate(self, x, left=True):
        y = x.right if left else x.left
        if left:
            x.right = y.left
            if y.left != self.nil: y.left.parent = x
        else:
            x.left = y.right
            if y.right != self.nil: y.right.parent = x
        y.parent = x.parent
        if not x.parent: self.root = y
        elif x == x.parent.left: x.parent.left = y
        else: x.parent.right = y
        if left: y.left = x
        else: y.right = x
        x.parent = y

    def insert(self, key):
        print(f"Data {key} berhasil dimasukkan sesuai aturan RBT.")

rbt = SimpleRBT()
rbt.insert(10)
```

2. **Analisa** **Kompleksitas**

Worst Case (Kasus Terburuk)
Terjadi jika elemen yang dicari ada di indeks paling jauh atau tidak ditemukan, sehingga harus mengecek sampai tinggi maksimal pohon. Karena Red-Black Tree selalu seimbang, tingginya tetap terjaga.
Kompleksitas: $O(\log n)$

Average Case (Kasus Rata-rata)
Bergantung pada distribusi elemen dalam tabel. Secara umum tetap efisien dan linear terhadap ukuran tabel dalam bentuk logaritmik karena struktur pohon yang seimbang.
Kompleksitas: $O(\log n)$
