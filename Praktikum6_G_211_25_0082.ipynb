```python
import math

def euclidean_distance(x, y):
    if len(x) != len(y):
        raise ValueError("Dimensi kedua vektor harus sama!")

    jumlah_kuadrat = 0
    for i in range(len(x)):
        jumlah_kuadrat += (x[i] - y[i]) ** 2

    return math.sqrt(jumlah_kuadrat)

titik_A = [1, 2, 3]
titik_B = [4, 5, 6]

jarak = euclidean_distance(titik_A, titik_B)
print("Hasil Jarak Euclidean:", jarak)
```
